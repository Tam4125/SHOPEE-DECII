```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/drive/MyDrive/AI Project/user_full_profile_with_segment.csv")
data.info()
```

```python
data = data.dropna()
data.info()
```

```python
loyal = data[data['cluster']==1]
potential = data[data['cluster']==2]
deal_hunter = data[data['cluster']==0]

loyal.shape, potential.shape, deal_hunter.shape
```

```python
# Sample rows
loyal_sample = loyal.sample(n=5000, random_state=42)
potential_sample = potential.sample(n=3000, random_state=42)
deal_sample = deal_hunter.sample(n=2000, random_state=42)

# Combine into one dataframe
new_df = pd.concat([loyal_sample, potential_sample, deal_sample], axis=0)

# Shuffle the final dataframe (optional)
new_df = new_df.sample(frac=1, random_state=42).reset_index(drop=True)

new_df.shape
```

```python
new_df['customer_segment'] = (
    new_df['customer_segment']
        .str.replace('Khach hang tiem nang', 'Potential')
        .str.replace('Khach hang VIP', 'Loyal')
)
```
